```python
import requests

STRAVA_AUTHORIZATION_PAGE_URL = 'https://www.strava.com/oauth/authorize'
CCDB_CLIENT_ID= '3869'
CCDB_REDIRECT_URL = 'http://closecalldatabase.com/strava-registration'
STRAVA_GET_AUTH_URL = 'https://www.strava.com/oauth/authorize?'

# The first call is a get to https://www.strava.com/oauth/authorize with the options you want,
# loaded as keyvalue paramaters

#example request for a user authorizatoin
#https://www.strava.com/oauth/authorize?client_id=9&response_type=code&redirect_uri=http://localhost/token_exchange.php&approval_prompt=force
"""
https://www.strava.com/oauth/authorize?
  client_id=9
  &response_type=code
  &redirect_uri=http://testapp.com/token_exchange
  &scope=write  # LEAVE BLANK FOR READONLY PERMISSIONS
  &state=mystate
  &approval_prompt=force
"""

payload = {
    'client_id': CCDB_CLIENT_ID,
    'response_type': 'code', # that is a strava constant
    'redirect_uri': CCDB_REDIRECT_URL,
}
r = requests.get(STRAVA_GET_AUTH_URL, params=payload)

print(r.url)

print(r.text)
```
